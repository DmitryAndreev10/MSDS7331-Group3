The data set that our group selected for this lab assignment is called the 'Online News Popularity Data Set' downloaded from the UCI Machine Learning Repository. The data consists of 39,644 news articles that were collected during 2013 - 2015 from www.Mashable.com, a popular social media and technology news website. Each article (instance) is described using 61 features ranging from simple measures such as the length of the title and the weekday on which it was published to more complex natural language processing measures such as the polarity of the positive and negative words. 

This feature set was created during the data collection and pre-processing stage of a study carried out by K. Fernandes et al. and is summarized in the 'Data Mining Type' section of this report. The purpose of the study is to predict the popularity of online news articles. Popularity in this case is defined as the number of times that the article is shared across all social media platforms.

With the advent of smartphones and social media putting an overwhelming amount of information at our fingertips, it is becoming exponentially more difficult to hold our attention on anything for more than 10 seconds. Long-form articles in the Times and 1000-words in-depth news stories are a thing of the past. Authors and news outlets alike are having to go through a lot of trial and error to figure out how to get the attention of their readers. This is where our prediction model comes into play. By understanding what variables have the greatest influence on the popularity of news articles, we can help these authors to cut down on this learning curve. Prior to publishing, we can evaluate whether the article will be popular or not and what can be done to make it more popular.

To tackle this problem, our goal is to create a prediction algorithm that has a binary output; whether an article is popular or not popular. We chose not to predict the number of shares that an article will have once published since that would require complex machine learning algorithms that are most likely outside the scope of this lab assignment. In addition, if the accuracy of the prediction model is low then it could potentially provide authors with false hope such as their article getting 1000 shares versus in reality it only yields 10 shares. Whereas the binary output doesn't promise something unrealistic but simply tells the author whether the article surpasses their chosen cutoff value for the number of shares. 

A good prediction model would classify whether an article is popular or not with a 73% accuracy rate based on the aforementioned study by K. Fernandes et al.
